## Visualizations of the data set

```python
import numpy as numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('flat-prices.csv')
data.head()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(y= 'town', data= data)
plt.show()

```

```python
plt.figure(figsize=(4,4))
sns.countplot(y= 'flat_type', data= data)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(y= 'storey_range', data= data)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(y= 'flat_model', data= data)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(y= 'lease_commence_date', data= data)
plt.show()
```

```python
plt.figure(figsize= (16,4))
sns.boxplot(x= 'resale_price', data= data)
plt.show()
```

```python
plt.figure(figsize= (16,4))
sns.boxplot(x= 'floor_area_sqm', data= data)
plt.show()
```
